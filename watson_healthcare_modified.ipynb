```python
def col_unique_value(d):
    for col in d:
        print(col,":", d[col].unique())
```

```python
col_unique_value(data)
```

```python
yes_no_col = ['OverTime','Attrition']
for i in yes_no_col:
    data[i].replace({"Yes":1,"No":0},inplace=True)
```

```python
col_unique_value(data)
```

```python
watson_data = pd.get_dummies(data = data, columns = ['Department','MaritalStatus','BusinessTravel','JobRole','EducationField'])
```

```python
watson_data
```

```python
label_encoder = LabelEncoder()
watson_data['Gender'] = label_encoder.fit_transform(watson_data['Gender'])
```

```python
col_unique_value(watson_data)
```

```python
col_to_scale = ["YearsWithCurrManager", "YearsSinceLastPromotion", "YearsInCurrentRole","YearsAtCompany","YearsAtCompany","WorkLifeBalance","TrainingTimesLastYear",
                "TotalWorkingYears","Shift","RelationshipSatisfaction","PercentSalaryHike","NumCompaniesWorked","MonthlyRate","MonthlyIncome","JobSatisfaction","JobLevel",
                "JobInvolvement","HourlyRate","EnvironmentSatisfaction","Education","DistanceFromHome","DailyRate","Age"]
scaler = MinMaxScaler()
watson_data[col_to_scale] = scaler.fit_transform(watson_data[col_to_scale])
```

```python
# col_unique_value(watson_data)
```

```python
watson_data.drop(["EmployeeCount","Over18","StandardHours"], axis =1, inplace=True)
```

```python
# col_unique_value(watson_data)
```

```python
X = watson_data.drop(["Attrition"], axis =1)
y = watson_data['Attrition']
y,X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
X_train.shape, y_train.shape , X_test.shape , y_test.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(45,input_shape=(45,),activation="softmax"),
    keras.layers.Dense(100,activation="relu"),
    keras.layers.Dense(1,activation="sigmoid")])
model.compile(optimizer="adam",
              loss = "binary_crossentropy",
              metrics =[ "accuracy"])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=10)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_predict = model.predict(X_test)
```
